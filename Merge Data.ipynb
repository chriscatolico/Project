## Merge Data
This Code merges the education data and zillow data to create the final database

```python
import pandas as pd
```

```python
educationdata = pd.read_csv('APIScores.csv')
zillowdata = pd.read_csv('ZillowData.csv')
```

```python
educationdata.head()
```

```python
zillowdata.head()
```

```python
data = pd.merge(educationdata, zillowdata, how='inner', left_on='ZipCode',right_on='name')
```

```python
finaldata = data[['ZipCode', 'Weighted 3 Year Average API','zindex','latitude3','longitude4',]]
```

```python
finaldata = finaldata.rename(columns={'zindex': 'Median Home Price','latitude3': 'Latitude','longitude4': 'Longitude'})
finaldata.head()
```

```python
finaldata['Weighted 3 Year Average API'] = finaldata['Weighted 3 Year Average API'].map("{:.2f}".format).astype('float')
```

```python
finaldata = finaldata.set_index('ZipCode')
```

```python
finaldata = finaldata.dropna(how='any')
```

```python
finaldata.to_csv('Final Data.csv')
```
