```python
import os
import pandas as pd
import csv
import sys
from config import google_api_key
import requests
```

```python
gkey = google_api_key
```

```python
path = "14avg.csv"
csv.field_size_limit(50000000)
file= pd.read_csv(path, encoding = "ISO-8859-1")
```

```python
for index, row in file.iterrows():
    if pd.isnull(file.iloc[index, 11]):
        target = file.iloc[index,8]
        target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
            'address={0}&key={1}').format(target, gkey)
        request = requests.get(target_url).json()
        try:
            if request['results'][0]['address_components'][4]['types'][0] == 'postal_code':                               
                file.iloc[index, 11] = request['results'][0]['address_components'][4]['long_name']
                print(f"updated record {request['results'][0]['address_components'][4]['long_name']}")
            elif request['results'][0]['address_components'][5]['types'][0] == 'postal_code':                               
                file.iloc[index, 11] = request['results'][0]['address_components'][5]['long_name']
                print(f"updated record {request['results'][0]['address_components'][5]['long_name']}")
            elif request['results'][0]['address_components'][6]['types'][0] == 'postal_code':                               
                file.iloc[index, 11] = request['results'][0]['address_components'][6]['long_name']
                print(f"updated record {request['results'][0]['address_components'][6]['long_name']}")
            elif request['results'][0]['address_components'][7]['types'][0] == 'postal_code':                               
                file.iloc[index, 11] = request['results'][0]['address_components'][7]['long_name']
                print(f"updated record {request['results'][0]['address_components'][7]['long_name']}")
            elif request['results'][0]['address_components'][8]['types'][0] == 'postal_code':                               
                file.iloc[index, 11] = request['results'][0]['address_components'][8]['long_name']
                print(f"updated record {request['results'][0]['address_components'][8]['long_name']}")
            else:
                file.iloc[index, 11] ='Unable to Find'
                print('Unable to Find')
        except IndexError:
             file.iloc[index, 11] = 'Unable to Find'
```

```python
file.to_csv('schooldata.csv',index=True)
```

```python
schooldata = pd.read_csv('schooldata.csv', encoding = "ISO-8859-1")
```

```python
schooldata = schooldata[schooldata.ZipCode != 'Unable to Find']
```

```python
schooldata = schooldata[schooldata['ZipCode'].map(len) == 5]
```

```python
schooldata = schooldata[schooldata['Type: E=Elementary School, M=Middle School, H=High School'] == 'H']
```

```python
schooldata = schooldata[['ZipCode', 'School Name', 'District Name', 'County Name', 'Number of Students Included in the 2013 Growth API', 'Weighted 3 Year Average API']]
```

```python
schooldata = schooldata.dropna(how='any')
```

```python
schooldata
```

```python
def weighted_score(x):
    names = {'weightedscore': (x['Number of Students Included in the 2013 Growth API'] * x['Weighted 3 Year Average API']).sum()/x['Number of Students Included in the 2013 Growth API'].sum()}
    return pd.Series(names, index=['weightedscore'])
```

```python
weightedscore = schooldata.groupby(['ZipCode']).apply(weighted_score)
```

```python
weightedscore.mean().reset_index().to_csv('APIScores.csv')
```
