```python
import requests
import json
import pprint
import pandas as pd
from config import api_key
from citipy import citipy
import random
import numpy as np
```

```python
maxlatitude, minlatitude = 49.3457868, 24.7433195 
maxlongitude, minlongitude = -66.9513812, -124.7844079
```

```python
cities = []
coordinates = []
i=0

while len(cities)<5:
    lat, long = (random.uniform(minlatitude,maxlatitude),random.uniform(minlongitude,maxlongitude))
    if citipy.nearest_city(lat, long).country_code == 'us':
        cities.append(citipy.nearest_city(lat, long))
        coordinates.append((lat,long))
        print(cities[i].city_name + " "+ cities[i].country_code)
        i+=1
```

```python
url = 'https://api.yelp.com/v3/businesses/search'
key = api_key

headers = {'Authorization': f'Bearer {key}'
          }
```

```python
params = []

for lat, long in coordinates:
    params.append({"categories": "hotdogs", #hotdogs is yelp's category keyword for fastfood
              'latitude': 37.8085393,
              'longitude': -122.4196538,
              'radius': '20000',
              'limit': 50,
              'sort_by': 'distance',
              'price':'1,2'
             })
```

```python
responses = []

responses.append(requests.get('https://api.yelp.com/v3/businesses/search', headers=headers, params = params[0]).json())
```

```python
pprint.pprint(responses)
```

```python
name = []
latitude = []
longitude = []
price = []
reviewcount = []
citynames = []

for i in range(0, len(responses)):
    for x in range(0,len(responses[i]['businesses'])):
        name.append(responses[i]['businesses'][x]['alias'])
        latitude.append(responses[i]['businesses'][x]['coordinates']['latitude'])
        longitude.append(responses[i]['businesses'][x]['coordinates']['longitude'])
        price.append(responses[i]['businesses'][x]['price'])
        reviewcount.append(responses[i]['businesses'][x]['review_count'])
        citynames.append(cities[i].city_name)
```

```python
df_restaurants = pd.DataFrame(
    {'Region': citynames,
     'Name': name,
     'Price': price,
     'Review Count': reviewcount,
     'latitude': latitude,
     'longitude': longitude
    }

)
```
