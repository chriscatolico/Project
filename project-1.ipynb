```python
import pandas as pd
import numpy as np

# Import the data into a Pandas DataFrame
scores_df = pd.read_csv("caa_ca2018_1_csv_v3.csv")
zip_df = pd.read_csv("caa_ca2018entities_csv.csv")
```

```python
scores_merged_df = pd.merge(scores_df, zip_df, on="School Code")
scores_merged_df.head()
```

```python
scores_merged_df.columns
```

```python
scores_clean_df = scores_merged_df.loc[:, ["School Code", "District Name", "County Name", "Zip Code", "Test Year_x", "Grade", "Test Id", "Students Tested",
                             "Mean Scale Score", "Percentage Level 3", "Percentage Level 2", "Percentage Level 1"]]
scores_clean_df.head()
```

```python
scores_clean_df["Zip Code"][0]
```

```python
scores_wzip_df = scores_clean_df.loc[(scores_clean_df["Zip Code"] != '     ')]
student_scores_df = scores_wzip_df.loc[(scores_clean_df["Students Tested"] != '*')]

df = student_scores_df.reset_index()
df.head()
```

```python
df[["Percentage Level 3", "Percentage Level 2", "Percentage Level 1"]] = df[["Percentage Level 3", "Percentage Level 2", "Percentage Level 1"]].apply(pd.to_numeric)
df.head()
```

```python
grouped_zip_df = df.groupby(["Zip Code", "Test Id"])
performance_zip_df = grouped_zip_df[["Percentage Level 3", "Percentage Level 2", "Percentage Level 1"]].mean()

performance_zip_df.reset_index()
```
